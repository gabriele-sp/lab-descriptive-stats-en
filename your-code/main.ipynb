#### Show the frequency distribution of the amount spent in Spain.

```python
# your answer here
# Calculate total amount spent
orders['AmountSpent'] = orders['Quantity'] * orders['UnitPrice']

# Filter for Spain
spain_orders = orders.loc[orders['Country'] == 'Spain']

# Plot histogram
plt.figure(figsize=(8,5))
sns.histplot(spain_orders['AmountSpent'], bins=20, kde=True, color='skyblue', edgecolor='black')
plt.title('Frequency Distribution of Amount Spent in Spain')
plt.xlabel('Amount Spent (€)')
plt.ylabel('Frequency')
plt.show()
```
